# Обоснование выбора диаграммы
Необходимо построить взаимосвязь суммарного числа оценок пользователей (для текущей версии) и средней оценки пользователей (для текущей версии). Заметим, что суммарное число оценок пользователей может повторяться. Например, для каких-то двух приложений может оказаться так, на них оставили по 10 оценок, но при этом оценки могут быть разные, и, как итог, у приложений будут разные средние оценки. В таком случае, диаграмма рассеивания наиболее удачный способ отразить взаимосвязь, так как она покажет тенденцию зависимости средней оценки пользователей от суммарного числа оценок с учетом того, что суммарное число оценок может повторяться
# Тенденции и взаимосвязи
Можно заметить, что с увеличением числа оценок растёт средняя оценка пользователей. То есть чем больше отзывов, тем точнее средняя оценка  
Также можно заметить, что суммарное число оценок в большей степени лежит между 0 и 40000  
Также можно заметить, что подавляющее число приложений имеет оценку 4.5 или 5.0

```python
# Вопрос 1: Как влияет возрастное ограничение контента (Content Rating) на средний размер приложения (size_bytes)? 

df.groupby("cont_rating")["size_bytes"].mean()

# По данной статистике можно увидеть, что приложения с ограничением контента 17+
# и 4+ имеют средний размер в 150МБ, а 9+ и 12+ имею средний размер почти в два раза выше
```

```python
# Вопрос 2: Какое количество скриншотов в описании (ipadSc_urls.num) наиболее часто встречается 
# у приложений с высоким средним рейтингом пользователей (user_rating)?

high_rating_apps = df[df['user_rating'] >= 4]
screen_counts = high_rating_apps['ipadSc_urls.num'].value_counts()

print(f"Наиболее часто встречаемое количество скриншотов: {screen_counts.idxmax()}")
print(f"Число приложений с этим количеством скриншотов: {screen_counts.max()}")

# Cначала фильтруем данные, чтобы оставить приложения с высоким рейтингом
# Группируем приложения по количеству скриншотов и считаем их количество
# Потом выводим информацию, какое количество скриншов встречается чаще
# и приложения с этим количеством скриншотов
```

```python
# Вопрос 3: Какие возрастные ограничения контента (Content Rating) наиболее распространены 
# среди приложений в определенных жанрах?

df.groupby(["prime_genre", "cont_rating"]).size().unstack(fill_value=0).idxmax(axis=1)

# Группируем данные по жанрам и возрастным ограничениям контента, подсчитываем их количество и
# выводим наиболее распостраненное возрастное ограничение контента для каждого жанра
```

```python
# Вопрос 4: Как распределены приложения по количеству скриншотов в описании (ipadSc_urls.num) в разных жанрах?
import matplotlib.pyplot as plt
genre_screen_counts = df.groupby("prime_genre")["ipadSc_urls.num"].value_counts().unstack(fill_value=0)

genre_screen_counts.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.xlabel("Жанр приложения")
plt.ylabel("Количество приложений")
plt.title("Распределение приложений по количеству скриншотов")
plt.legend(title="Количество скриншотов", loc="upper right")
plt.xticks(rotation=90)
plt.show()

# По диаграмме видно, что жанр приложений Games имеет наиболее высокое количество
# скриншотов по сравнению с другими жанрами. 
```

```python
# Вопрос 5: Есть ли различия в цене приложений в зависимости от поддерживаемых устройств (sup_devices.num)?

avg_price_devices = df.groupby("sup_devices.num")["price"].mean()

avg_price_devices.plot(kind='bar', figsize=(10, 6))
plt.xlabel("Количество поддерживаемых устройств")
plt.ylabel("Средняя цена приложения")
plt.title("Средняя цена приложений в зависимости от поддерживаемых устройств")
plt.xticks(rotation=0)
plt.show()

# Можно заметить, что средняя цена значительно выше у приложений, которые поддерживают
# 16 устройств. Остальные приложеения не выбиваются из статистики.
```
